```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularSinglyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def print_list(self):
        if self.head is None:
            print("List is empty.")
            return
        temp = self.head
        nodes = []
        while True:
            nodes.append(temp.data)
            temp = temp.next
            if temp == self.head:
                break
        print(" -> ".join(map(str, nodes)))

    def pop_first(self):
        if self.head is None:
            print("List is empty.")
            return None

        if self.head.next == self.head:  # Only one node
            data = self.head.data
            self.head = None
            return data

        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        data = self.head.data
        self.head = self.head.next
        temp.next = self.head
        return data

    def pop(self, index):
        if self.head is None:
            print("List is empty.")
            return None

        if index == 0:
            return self.pop_first()

        temp = self.head
        prev = None
        count = 0

        while True:
            if count == index:
                data = temp.data
                prev.next = temp.next
                return data
            prev = temp
            temp = temp.next
            count += 1
            if temp == self.head:
                print("Index out of bounds.")
                return None

    def remove(self, value):
        if self.head is None:
            print("List is empty.")
            return

        temp = self.head
        prev = None
        while True:
            if temp.data == value:
                if prev is None:  # Removing head
                    return self.pop_first()
                else:
                    prev.next = temp.next
                    if temp == self.head:  # Updating head if necessary
                        self.head = prev.next
                    return
            prev = temp
            temp = temp.next
            if temp == self.head:
                print("Value not found in the list.")
                return

    def delete_all_nodes(self):
        if self.head is None:
            print("List is empty.")
            return

        self.head = None
        print("All nodes have been deleted.")

# Example usage
csll = CircularSinglyLinkedList()

# Append nodes
csll.append(10)
csll.append(20)
csll.append(30)
csll.append(40)

# Print the list
print("Circular Singly Linked List:")
csll.print_list()  # Output: 10 -> 20 -> 30 -> 40

# Pop the first node
print("Popped first node:", csll.pop_first())  # Output: 10
csll.print_list()  # Output: 20 -> 30 -> 40

# Pop node at index 1
print("Popped node at index 1:", csll.pop())  # Output: 30
csll.print_list()  # Output: 20 -> 40

# Remove node with value 40
csll.remove(40)
print("After removing 40:")
csll.print_list()  # Output: 20

# Delete all nodes
csll.delete_all_nodes()
csll.print_list()  # Output: List is empty.
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularSinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
            new_node.next = self.head
        else:
            self.tail.next = new_node
            self.tail = new_node
            self.tail.next = self.head

    def print_list(self):
        if not self.head:
            print("List is empty")
            return
        current = self.head
        while True:
            print(current.data, end=" -> ")
            current = current.next
            if current == self.head:
                break
        print()

    def pop(self):
        if not self.head:
            print("List is empty. Nothing to pop.")
            return

        if self.head == self.tail:  # Only one element
            popped_data = self.head.data
            self.head = None
            self.tail = None
            return popped_data

        # More than one element
        popped_data = self.tail.data
        self.tail = self.tail.next  # Move tail to the previous node
        self.head.next = self.tail.next  # Maintain the circular link
        return popped_data


# Example usage:
csll = CircularSinglyLinkedList()
csll.append(10)
csll.append(20)
csll.append(30)
csll.print_list()

print("Popped element:", csll.pop())
csll.print_list()

print("Popped element:", csll.pop())
csll.print_list()

print("Popped element:", csll.pop())
csll.print_list()
```

Node Class: Defines the structure of a node in the linked list with data and next attributes.

CircularSinglyLinkedList Class:

append Method: Adds a new node to the end of the list and maintains the circular link.

print_list Method: Traverses and prints the circular linked list.

pop Method:
Checks if the list is empty.
If the list contains only one node, it removes it and sets head and tail to None.

If the list contains more than one node, it removes the last node (tail) by moving the tail pointer to the previous node and maintaining the circular link.

Key Points:
Big O(1) Time Complexity: The pop method is designed to execute in constant time ùëÇ(1)
O(1), which means the operation is independent of the size of the linked list.

Efficient Tail Management: 
By keeping track of the tail, we efficiently update the list structure when the last element is removed.

This method efficiently removes the last element while maintaining the circular structure of the linked list.

To optimize the methods for pop, remove, and delete_all_nodes in a 
Circular Singly Linked List (CSLL) to have ùëÇ(1)

O(1) time complexity, you need to focus on the following:

    Maintain a Tail Pointer: Keep track of the last node (tail) to efficiently manage the list.

    Efficiently Update Pointers: 
    Ensure that operations do not require traversal of the list, which would increase time complexity.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularSinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
            new_node.next = self.head
        else:
            self.tail.next = new_node
            self.tail = new_node
            self.tail.next = self.head

    def pop_first(self):
        if self.head is None:
            print("List is empty.")
            return None

        data = self.head.data

        if self.head == self.tail:  # Only one element
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            self.tail.next = self.head  # Maintain circular link

        return data

    def pop(self, index):
        if self.head is None:
            print("List is empty.")
            return None

        if index == 0:
            return self.pop_first()

        temp = self.head
        prev = self.tail  # Start with the tail
        count = 0

        while count < index:
            prev = temp
            temp = temp.next
            count += 1
            if temp == self.head:
                print("Index out of bounds.")
                return None

        data = temp.data
        prev.next = temp.next
        if temp == self.tail:
            self.tail = prev  # Update tail if last node is removed
        return data

    def remove(self, value):
        if self.head is None:
            print("List is empty.")
            return

        temp = self.head
        prev = self.tail  # Start with the tail

        while True:
            if temp.data == value:
                if temp == self.head:
                    return self.pop_first()
                prev.next = temp.next
                if temp == self.tail:
                    self.tail = prev  # Update tail if last node is removed
                return
            prev = temp
            temp = temp.next
            if temp == self.head:
                print("Value not found in the list.")
                return

    def delete_all_nodes(self):
        self.head = None
        self.tail = None
        print("All nodes have been deleted.")

# Example usage:
csll = CircularSinglyLinkedList()
csll.append(10)
csll.append(20)
csll.append(30)
csll.append(40)

print("Popped element at index 2:", csll.pop(2))
csll.remove(20)
csll.delete_all_nodes()
```

Key Improvements:
1.	Tail Pointer Management:
	tail pointer is maintained to keep track of the last node in the list, allowing us to update the list in O(1)O(1)O(1) time.
2.	pop_first Method:
	Removes the first element and updates the head and tail pointers accordingly.
	If there's only one element, both head and tail are set to None.
3.	pop Method:
	Directly accesses the node to be removed by maintaining a reference to the previous node (prev), and adjusts pointers in O(1)O(1)O(1) time.
	Handles the removal of the tail node specifically.
4.	remove Method:
	Removes the node with the given value, updating the tail if necessary.
	Maintains O(1)O(1)O(1) time complexity by using the tail pointer.
5.	delete_all_nodes Method:
	Efficiently clears the entire list by setting head and tail to None.
