Doubly Linked List

    In a doubly linked list, each node contains:

    Data: The value of the node.
    Next: A pointer/reference to the next node in the list.
    Prev: A pointer/reference to the previous node in the list.

Node Class Constructor

    Define a Node class that includes 
    the data and pointers to both 
    the next and previous nodes.

```python
class Node:
     def __init__(self, data):
        self.data = data       # Data stored in the node
        self.next = None       # Pointer to the next node
        self.prev = None       # Pointer to the previous node
        print("Node is ",data)

Node1 = Node(100)
print(Node1.next)
print(Node1.prev)
```

Append Method in Doubly Linked List

      Adds a new node at the end of the list. 
      In a doubly linked list, each node has a reference to both 
      the next node and the previous node, 
      which makes appending a node more straightforward compared to a singly linked list.

Append Method Steps

Create a New Node: 
Allocate memory for the new node and set its data.

Check if the List is Empty: 

If the list is empty (i.e., the head is None), 
set the new node as the head of the list.

Traverse to the End of the List: 

If the list is not empty, 
start from the head and 
traverse to the last node.

Update Pointers:
Set the next pointer of the last node to the new node.

Set the prev pointer of the new node to the last node.

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None # Initialisation of head node
        self.tail = None # Initialisation of tail node

    def append(self, data):
        new_node = Node(data) # new_node.data = 100
        if self.head is None:  # If the list is empty For First Node
            self.head = new_node
            print("Head node data",self.head.data)  # 100
            self.tail = new_node
            print("tail node data",self.tail.data)
        else:
            self.tail.next = new_node  # Link the new node to the end of the list
            print("Tail Next node",self.tail.next.data) # 200
            new_node.prev = self.tail  # Link the new node back to the current tail
            print("Previous Node",new_node.prev.data)
            self.tail = new_node       # Update the tail to the new node
            print("Updated tail data",self.tail.data)

    def print_list(self):
        current = self.head
        print("DoublyLinkedList:")
        while current:
            print(current.data,end='<->')
            current = current.next
        print("None")

dll = DoublyLinkedList()
dll.append(100)
dll.append(200)
dll.append(300)
dll.append(400)
dll.print_list()
```

To give a practical example, 

letâ€™s consider a doubly linked list initially containing 

the elements 100 and 200. We append 300 to this list:

Initial List: 100 <-> 200

New Node: 300

Traverse to the end of the list (200).

Update the next pointer of 200 to 300

Set the prev pointer of 300 to 200

Final List: 100 <-> 200 <-> 300

Time Complexity

    Traversal to End:

    If a tail pointer is maintained, this step is O(1) because 
    we can directly access the tail node without traversal.

    If no tail pointer is maintained, the traversal would take O(n), 
    where n is the number of nodes in the list.

Insertion: O(1)

    Once the correct position (end) is found, inserting the node is done in constant time.

Overall Time Complexity:

    With Tail Pointer: O(1)

    Without Tail Pointer: O(n)

Space Complexity

    New Node Creation: O(1)

    Only a single new node is created, regardless of the list's size.

    Overall Space Complexity: O(1)

__str__ Method for Doubly Linked List


        The __str__ method provides a way to print the doubly linked list in a readable format.

Algorithm

    Initialize an Empty List: Create an empty list to collect node data.
    Start from the Head: Begin at the head of the doubly linked list.

    Traverse the List:

    For each node, append its data (converted to a string) to the list.
    Move to the next node.

    Join the Collected Data: Join the collected data with a separator (e.g., " <-> ").
    Return the Result: Return the resulting string.

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:  # If the list is empty
            self.head = new_node
            print("Head node data",self.head.data)
            self.tail = new_node
            print("tail node data",self.tail.data)
        else:
            self.tail.next = new_node  # Link the new node to the end of the list
            print("Tail Next node",self.tail.next.data)
            new_node.prev = self.tail  # Link the new node back to the current tail
            print("Previous Node",new_node.prev.data)
            self.tail = new_node       # Update the tail to the new node
            print("Updated tail data",self.tail.data)

    def __str__(self):
        nodes = []
        current = self.head
        while current:
            nodes.append(str(current.data))
            current = current.next
        return " <-> ".join(nodes)


dll = DoublyLinkedList()
dll.append(100)
dll.append(200)
dll.append(300)
dll.append(400)
print("Doubly Linked List is ",dll)
```

Time Complexity

    Traversal: O(n), where n is the number of nodes in the list. Each node is visited once.

    String Joining: O(n) for creating the final string from the list of node data.
    Overall Time Complexity: O(n)

    Space Complexity
    Node Data List: O(n) for storing the data of n nodes.
    Resulting String: O(n) for the final joined string.
    Overall Space Complexity: O(n)

Prepend Method in Doubly Linked List

    The prepend method adds a new node to the beginning of a doubly linked list.
    

Algorithm

    Create a New Node: Initialize a new node with the given data.

    Check if List is Empty:

    If the list is empty (self.head is None), set the new node as the head.

    Insert at the Beginning:

    If the list is not empty:

    Set the next pointer of the new node to the current head.

    Set the prev pointer of the current head to the new node.

    Update the head to be the new node.

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:  # If the list is empty
            self.head = new_node
            print("Head node data",self.head.data)
            self.tail = new_node
            print("tail node data",self.tail.data)
        else:
            self.tail.next = new_node  # Link the new node to the end of the list
            print("Tail Next node",self.tail.next.data)
            new_node.prev = self.tail  # Link the new node back to the current tail
            print("Previous Node",new_node.prev.data)
            self.tail = new_node       # Update the tail to the new node
            print("Updated tail data",self.tail.data)

    def prepend(self, data):
        new_node = Node(data)
        if self.head is None:  # If the list is empty
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head  # Link the new node to the start of the list
            print("Next Node ",new_node.next.data)
            self.head.prev = new_node  # Link the old head back to the new node
            print("Previous node",self.head.prev.data)
            self.head = new_node       # Update the head to the new node
            print("Updated head data",self.head.data)

    def reverse_traverse(self):
        current = self.head
        # Traverse to the end of the list
        while current and current.next:
            current = current.next
        # Traverse backwards
        while current:
            print(current.data, end='<->')
            current = current.prev
        print()

    def reverse_traverse1(self):
        current = self.tail
        # Traverse backwards
        while current:
            print(current.data, end='<->')
            current = current.prev
        print()

    def __str__(self):
        nodes = []
        current = self.head
        while current:
            nodes.append(str(current.data))
            current = current.next
        return " <-> ".join(nodes)


dll = DoublyLinkedList()
dll.append(100)
dll.append(200)
dll.append(300)
dll.append(400)
print("Doubly Linked List is ",dll)

dll.prepend(50)

print("Doubly Linked List after prepend ",dll)
print("Reverse Traverse1")
dll.reverse_traverse1()
```

Reverse Traverse Method for Doubly Linked List

Algorithm
Traverse to the End: 
Start from the head and move to the last node by following the next pointers.

Iterate Backwards:

From the last node, perform the desired operation (e.g., print the data).

Move to the previous node using the prev pointer.

Continue Until Start: Continue this process until reaching the head 
(current becomes None).

```python
def reverse_traverse(self):
        current = self.head
        # Traverse to the end of the list
        while current and current.next:
            current = current.next
        # Traverse backwards
        while current:
            print(current.data, end=' ')
            current = current.prev
        print()
```

Search Method

    The search method looks for a node with a specified value in the doubly linked list.

Algorithm

    Start from the Head: Begin at the head of the list.

Iterate Through the List:

    Compare the value of the current node with the target value.
    If a match is found, return the node or its position.
    Move to the next node using the next pointer.
    
Continue Until End:

    If the end of the list is reached (current becomes None), 
    return None or indicate that the value was not found.


```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:  # If the list is empty
            self.head = new_node
            print("Head node data",self.head.data)
            self.tail = new_node
            print("tail node data",self.tail.data)
        else:
            self.tail.next = new_node  # Link the new node to the end of the list
            print("Tail Next node",self.tail.next.data)
            new_node.prev = self.tail  # Link the new node back to the current tail
            print("Previous Node",new_node.prev.data)
            self.tail = new_node       # Update the tail to the new node
            print("Updated tail data",self.tail.data)

    def search(self, value):
        current = self.head
        while current:
            if current.data == value:
                print("Value Address is",id(current))  # Or return position if needed
            current = current.next
        return ValueError

    def __str__(self):
        nodes = []
        current = self.head
        while current:
            nodes.append(str(current.data))
            current = current.next
        return " <-> ".join(nodes)


dll = DoublyLinkedList()
dll.append(100)
dll.append(200)
dll.append(300)
dll.append(400)
print("Doubly Linked List is ",dll)
dll.search(300)
```

Get Method

    The get method retrieves the value of a node at a specified position (index) in the doubly linked list.

Algorithm

    1. Start from the Head: Begin at the head of the list.

    2. Iterate to Position:

        Move through the list by following the next pointers until reaching the specified index.

    3. Return the value of the node at that position.

    4. Check for Valid Index: 
        If the index is out of bounds (i.e., the end of the list is reached before the index), 
        handle this case appropriately (e.g., return None or raise an IndexError).


```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:  # If the list is empty
            self.head = new_node
            print("Head node data",self.head.data)
            self.tail = new_node
            print("tail node data",self.tail.data)
        else:
            self.tail.next = new_node  # Link the new node to the end of the list
            print("Tail Next node",self.tail.next.data)
            new_node.prev = self.tail  # Link the new node back to the current tail
            print("Previous Node",new_node.prev.data)
            self.tail = new_node       # Update the tail to the new node
            print("Updated tail data",self.tail.data)

    def search(self, value):
        current = self.head
        while current:
            if current.data == value:
                print("Value Address is",id(current))  # Or return position if needed
            current = current.next
        return ValueError

    def get(self, index):
        current = self.head #start with head
        count = 0 # Initialise counter
        while current:
            if count == index:  # if Index found
                return current.data # return data by index
            count += 1          # Increment counter till given index position
            current = current.next
        return IndexError  # Index out of bounds

    def __str__(self):
        nodes = []
        current = self.head
        while current:
            nodes.append(str(current.data))
            current = current.next
        return " <-> ".join(nodes)


dll = DoublyLinkedList()
dll.append(100)
dll.append(200)
dll.append(300)
dll.append(400)
print("Doubly Linked List is ",dll)
dll.search(300)
dll.get(1)
```

Overall Time Complexity: O(n)

Temporary Variables: O(1) - Only a few variables are used (current node pointer and counter).
Overall Space Complexity: O(1)

Set Method

    The set method updates the value of a node at a specified position (index) in the doubly linked list.

Algorithm
    Start from the Head: Begin at the head of the list.

    Iterate to Position:
        Move through the list by following the next pointers until reaching the specified index.
        Update the value of the node at that position with the new value.

    Check for Valid Index: 
        If the index is out of bounds (i.e., the end of the list is reached before the index), 
        handle this case appropriately (e.g., return False or raise an IndexError).

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:  # If the list is empty
            self.head = new_node
            print("Head node data",self.head.data)
            self.tail = new_node
            print("tail node data",self.tail.data)
        else:
            self.tail.next = new_node  # Link the new node to the end of the list
            print("Tail Next node",self.tail.next.data)
            new_node.prev = self.tail  # Link the new node back to the current tail
            print("Previous Node",new_node.prev.data)
            self.tail = new_node       # Update the tail to the new node
            print("Updated tail data",self.tail.data)

    def search(self, value):
        current = self.head
        while current:
            if current.data == value:
                print("Value Address is",id(current))  # Or return position if needed
            current = current.next
        return ValueError

    def get(self, index):
        current = self.head
        count = 0
        while current:
            if count == index:
                return current.data
            count += 1
            current = current.next
        return IndexError  # Index out of bounds

    def set(self, index, value):
        current = self.head
        count = 0
        while current:
            if count == index:  # match counter
                current.data = value # set new value
                print("Updated Value by index",current.data)
            count += 1 # increment counter till given index position
            current = current.next # update current pointer
        return IndexError  # Index out of bounds

    def __str__(self):
        nodes = []
        current = self.head
        while current:
            nodes.append(str(current.data))
            current = current.next
        return " <-> ".join(nodes)


dll = DoublyLinkedList()
dll.append(100)
dll.append(200)
dll.append(300)
dll.append(400)
print("Doubly Linked List is ",dll)
dll.search(300)
dll.get(1)
dll.set(1,150)
dll.get(1)
print("Doubly Linked After set List is ",dll)
```

Insert Method:
    The insert method adds a new node at a specified position (index) in the doubly linked list.

Algorithm

1. Create a New Node: Initialize a new node with the given data.

2. Check for Empty List:

3. If inserting at index 0 (the head), call the prepend method.

4. Traverse to the Position:

    Start from the head and move to the node just before the specified index.

5. Insert the New Node:
    Update the next pointer of the new node to point to the current node at the index.

    Update the prev pointer of the new node to point to the previous node.

    Update the next pointer of the previous node to point to the new node.

    Update the prev pointer of the current node to point to the new node.


```python
#%%writefile DoublyLinkedListModule.py

class Node:
     def __init__(self, data):
        self.data = data       # Data stored in the node
        self.next = None       # Pointer to the next node
        self.prev = None       # Pointer to the previous node


class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:  # If the list is empty
            self.head = new_node
            print("Head node data",self.head.data)
            self.tail = new_node
            print("tail node data",self.tail.data)
        else:
            self.tail.next = new_node  # Link the new node to the end of the list
            print("Tail Next node",self.tail.next.data)
            new_node.prev = self.tail  # Link the new node back to the current tail
            print("Previous Node",new_node.prev.data)
            self.tail = new_node       # Update the tail to the new node
            print("Updated tail data",self.tail.data)

    def insert(self, index, data):
        new_node = Node(data)
        if index == 0:
            self.prepend(data)
            return
        current = self.head
        count = 0
        while current and count < index - 1:
            current = current.next
            count += 1
        if current is None:
            return  # Index out of bounds
        new_node.next = current.next
        if current.next:
            current.next.prev = new_node
        current.next = new_node
        new_node.prev = current


    def search(self, value):
        current = self.head
        while current:
            if current.data == value:
                print("Value Address is",id(current))  # Or return position if needed
            current = current.next
        return ValueError

    def get(self, index):
        current = self.head
        count = 0
        while current:
            if count == index:
                return current.data
            count += 1
            current = current.next
        return IndexError  # Index out of bounds

    def pop_first(self):
        if self.head is None:
            return None  # List is empty
        removed_node = self.head
        self.head = self.head.next
        if self.head:
            self.head.prev = None
        removed_node.next = None
        return removed_node.data

    def pop(self):
        if self.head is None:
            return None  # List is empty
        current = self.head
        while current.next:
            current = current.next
        if current.prev:
            current.prev.next = None
        else:
            self.head = None
        return current.data

    def remove(self, data):
        current = self.head
        while current:
            if current.data == data:
                if current.prev:
                    current.prev.next = current.next
                else:
                    self.head = current.next
                if current.next:
                    current.next.prev = current.prev
                else:
                    self.tail = current.prev
                return True
            current = current.next
        return False

    def __str__(self):
        nodes = []
        current = self.head
        while current:
            nodes.append(str(current.data))
            current = current.next
        return " <-> ".join(nodes)

if __name__ == '__main__':
    dll = DoublyLinkedList()
    dll.append(100)
    dll.append(200)
    dll.append(300)
    dll.append(400)
    print("Doubly Linked List is ",dll)
    dll.search(300)
    dll.get(1)
    dll.insert(2,250)
    print(dll)
    #dll.pop()
    dll.remove(100)
    print("DLL after remove",dll)
```

Remove Method
The remove method deletes the first occurrence of a node with the specified value.

Algorithm
Start from the Head: Begin at the head of the list.
Search for Node:
Compare the value of each node with the target value.
If a match is found, perform the removal.
Update Pointers:
If the node to be removed is the head, update the head to the next node.
Update the prev pointer of the next node, if it exists.
Update the next pointer of the previous node, if it exists.
Remove Reference:
Optionally, set the removed nodeâ€™s pointers to None to help with garbage collection.

```python
class Node:
     def __init__(self, data):
        self.data = data       # Data stored in the node
        self.next = None       # Pointer to the next node
        self.prev = None       # Pointer to the previous node


class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:  # If the list is empty
            self.head = new_node
            print("Head node data",self.head.data)
            self.tail = new_node
            print("tail node data",self.tail.data)
        else:
            self.tail.next = new_node  # Link the new node to the end of the list
            print("Tail Next node",self.tail.next.data)
            new_node.prev = self.tail  # Link the new node back to the current tail
            print("Previous Node",new_node.prev.data)
            self.tail = new_node       # Update the tail to the new node
            print("Updated tail data",self.tail.data)

    def remove(self, data):
        '''Remove given data'''
        current = self.head
        if current.data == data:  # first element
            self.head = None # initialisation
        while current:
            if current.data == data:
                if current.prev:
                    current.prev.next = current.next # update pointer of current node
                else:
                    self.head = current.next
                    print("Head node pointer updated to next node")
                if current.next:
                    current.next.prev = current.prev
                    print("Update pointer")
                else:
                    self.tail = current.prev
                    print("Updated Tail Pointer to previous node")
                return True
            else:
                current = current.next
                print("Update current node pointer address")
                print("Removed data is ",current.data)
        return False

    def __str__(self):
        nodes = []
        current = self.head
        while current:
            nodes.append(str(current.data))
            current = current.next
        return " <-> ".join(nodes)

dll = DoublyLinkedList()
dll.append(100)
dll.append(200)
dll.remove(200)
print(dll)
```
